```python
#creating a custom function to retrieve the emotion label for all the individual rows of the dataframe
```

```python
def get_emotion_label(text):
    return (emotion(text)[0]['label'])
```

```python
#example
```

```python
get_emotion_label('I am excited about todays match')
```

```python
text_df['emotion']=text_df['content'].apply(get_emotion_label)
```

```python
text_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(16,8))
cp=sns.countplot(x='emotion',data=text_df,order=text_df['emotion'].value_counts(ascending=False).index)
cp.bar_label(cp.containers[0])
plt.xticks(rotation ='vertical')
plt.show()
```
